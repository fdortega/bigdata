Parallelize the following codes using Dask's @delayed decorator and derive the task execution graph for each.
1. 

```python
# Import
from dask import delayed

@delayed
def sum_up(l):
    return sum(l)

@delayed
def cube(n):
    return n**3
```

```python
total = []
for i in range(1,10):
    total.append(cube(i))

result_cube = (sum_up(total))
result_cube
```

```python
# Visualize task execution graph
result_cube.visualize()
```

```python
# Compute actual value
result_cube.compute()
```

2. 

```python
# first two fibonacci numbers
fibonacci_nums = [1,1]

@delayed
# returns the nth fibonacci number
def get_nth_fibonacci_number(n):
    if n == 1:
        return fibonacci_nums[0]
    elif n == 2:
        return fibonacci_nums[1]
    else:
        return get_nth_fibonacci_number(n-1).compute() + get_nth_fibonacci_number(n-2).compute()

result_fib = (get_nth_fibonacci_number(10))
result_fib
```

```python
# Visualize task execution graph
result_fib.visualize()
```

```python
# Compute actual value
result_fib.compute()
```
